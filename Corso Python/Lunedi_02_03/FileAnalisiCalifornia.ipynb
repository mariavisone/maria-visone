import numpy as np # Coordinate di San Francisco 
sf_lat, sf_long = 37.77, -122.42 
# Calcolo distanza approssimativa (in gradi) df["dist_sf"] = np.sqrt((df["Latitude"] - sf_lat)2 + (df["Longitude"] - sf_long)2) # Feature combinata: reddito ponderato per distanza df["MedInc_x_distSF"] = df["MedInc"] / (df["dist_sf"] + 0.01) # +0.01 per evitare divisione per zero
